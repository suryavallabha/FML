```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
def tanh(x):
 return np.tanh(x)
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :tanh')
plt.show()
 
```
